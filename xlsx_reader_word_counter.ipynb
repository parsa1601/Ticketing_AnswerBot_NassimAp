```python


with open ('shitty.txt', encoding="utf8") as shitty:
    # shit = shitty.readlines()
    # shit = list(shitty)
    shit = shitty.read().split('\n')


with open ('worthy.txt', encoding="utf8") as worthy:
    worth = worthy.read().split('\n')

shit
```

```python
       
from collections import OrderedDict

txt1 = description['description']
txt3 = txt1.to_list()
txt2 = comments['commentcontent']
txt4 = txt2.to_list()
txt3.extend(txt4)
main_span = 2

dictionary = dict()

# print(txt3[0])
for txt in txt3:
    words = txt.split()
    new_words = []
  
           
    for index , elem in enumerate(words):
        if (elem == 'می'): 
            words[index] = " "
            words[index+1] = "می" + words[index+1]
            words.pop(index)
            
        if (elem == 'نمی'): 
            words[index] = " "
            words[index+1] = "نمی" + words[index+1]
            words.pop(index)
        
        
    span = main_span
    for i in range (0, len(words), span):
        word = " ".join(words[i:i+span])
        new_words.append(word)
        
    for x in new_words:
         
        
        if(dictionary.__contains__(x)):
             dictionary[x] = dictionary[x] + 1           
        else:
            dictionary.update({x:1})
         
        

type(dictionary.keys())
# dictionary
```

```python
def weight_cal(words, count_of_elements, ww):
    if (count_of_elements == 2):
        
        for w in worth:    
            if words[0] == w:
                ww += 3
            elif words[1] == w:
                ww += 3
        for s in shit:
            if words[0] == s:
                # print("the word0 is: " , words[0] , " the shit is: " , s)     
                ww -= 3
            elif words[1] == s:
                # print("the word1 is: " , words[1] , " the shit is: " , s) 
                ww -= 3
                
    elif (count_of_elements == 1):
        
        for w in worth:    
            if words[0] == w:
                ww += 3    
        for s in shit:
            if words[0] == s:
                ww -= 3
    
    # print("weigth in function: " , ww)
    return ww
    
```
